# Deal with outliers

```python
dataset = data.iloc[:,:-1]
dataset.boxplot( rot = 90, figsize=(25,10))
```

```python
outlier_variables = ['CustomerID', 'Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in outlier_variables:
        q1=data[i].quantile(0.25)
        q3=data[i].quantile(0.75)
        iqr=q3-q1
        upper=q3+1.5*iqr
        lower=q1-1.5*iqr
        data[i]=np.where(data[i] >upper, upper, data[i])
        data[i]=np.where(data[i] <lower, lower, data[i])
```

```python
dataset = data.iloc[:,:-1]
dataset.boxplot( rot = 90, figsize=(25,10))
```

# SCALING

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
dataset = data.iloc[:,2:]
dataset = mms.fit_transform(dataset)
```

```python
dataset
```

# Kmeans clustering algorithm

```python
from sklearn.cluster import KMeans
```

```python
TWSS = []
k = list(range(2,9))

for i in k:
    kmeans = KMeans(n_clusters=i, init = 'k-means++')
    kmeans.fit(dataset)
    TWSS.append(kmeans.inertia_)
```

```python
TWSS
```

```python
plt.plot(k,TWSS,'ro--')
plt.xlabel('no of cluster')
plt.ylabel('TWSS')
```

```python
model = KMeans(n_clusters=4)
model.fit(dataset)
```

```python
mb = pd.Series(model.labels_)
```

```python
data['clust'] = mb
```

```python
data.head()
```

```python
data.tail()
```

# Add the cluster data with primary dataset

```python
data.to_csv('Kmeans_cluster.csv', encoding='utf-8')
```

# Split the data into dependent and independent variables

```python
dataset = pd.read_csv('Kmeans_cluster.csv')
```

```python
dataset.head()
```

```python
dataset.drop(columns=['Unnamed: 0','CustomerID'],inplace=True)
```

```python
dataset.head()
```

```python
sns.countplot(dataset['clust'] )
dataset['clust'].value_counts()
```

```python
sns.pairplot(dataset.iloc[:,:])
```

```python
dataset['Gender'].replace({'Male' : 1,'Female' : 0 }, inplace=True)
```

```python
X = dataset.iloc[:,:-1]
y = dataset['clust']
```

```python
X
```

```python
y
```

# Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

# Test the model

```python
log_pred = reg.predict(X_test)
log_pred
```

# Measure the performance using evaluation metrics

```python
from sklearn import metrics
metrics.confusion_matrix(y_test,log_pred)
```

```python
metrics.accuracy_score(y_test, log_pred)
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand = RandomForestClassifier()
rand.fit(X_train,y_train)
```

```python
rand_pred = rand.predict(X_test)
rand_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test, rand_pred)
```

```python
confusion_matrix(y_test,rand_pred)
```

# Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
dtree_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, dtree_pred)
```

```python
confusion_matrix(y_test, dtree_pred)
```

# KNeighbors classifier
